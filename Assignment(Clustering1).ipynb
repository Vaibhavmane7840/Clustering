#Kmeans_Clustering

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans 
```

```python
crime_data = pd.read_csv("crime_data.csv")#read dataset
crime_data.head()
```

```python
#normalize data
from sklearn.preprocessing import MinMaxScaler
data = MinMaxScaler()#standardization
data  = pd.DataFrame(data.fit_transform(crime_data.iloc[:,1:]))
data
```

```python
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(data)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()  
```

```python
clf = KMeans(n_clusters=4)
y_kmeans = clf.fit_predict(data)
y_kmeans
```

```python
clf.cluster_centers_
```

```python
clf.inertia_
```

```python
md = pd.Series(y_kmeans) # converting numpy array into pandas series object 
crime_data['clus']= md# creating a  new column and assigning it to new column
crime_data
```

```python
crime_data.iloc[:,1:].groupby(crime_data['clus']).mean()
```

```python
plt.style.use(['classic']) 
print(plt.style.available)
```

```python
crime_data.plot(x="Murder",y ="Assault",c=clf.labels_,kind="scatter",s=30 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans')
```

#DBSCAN CLUSTERING

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
crime_data = pd.read_csv("crime_data.csv")
crime_data.head(6)
```

```python
crime_data.info()
```

```python
data = crime_data.iloc[:,1:]
array = data.values
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array) 
X   
```

```python
dbscan = DBSCAN(eps=2, min_samples=5)
dbscan.fit(X) 
```

```python
#Noisy samples are given the label -1.
dbscan.labels_ 
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```
